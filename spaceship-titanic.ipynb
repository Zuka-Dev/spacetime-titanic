```python
#Import files
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("./train.csv")
df.head(n=10)
```

```python
destinations = pd.get_dummies(df.Destination)
destinations
```

```python
newDF = pd.concat([df, destinations], axis="columns")
```

```python
new2DF = newDF.drop(['Destination','55 Cancri e'], axis="columns")
new2DF
```

```python
new3DF = new2DF.dropna(axis=0)
new2DF.shape
```

```python
new3DF.shape
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
X_train = new3DF.drop(new3DF[["PassengerId","HomePlanet","Cabin","RoomService","FoodCourt","ShoppingMall","Spa","Name","VRDeck","Transported"]], axis="columns")
X_train.head()
```

```python
y_train = new3DF["Transported"]
y_train.head()
```

# Training model

```python
model.fit(X_train, y_train)
```
